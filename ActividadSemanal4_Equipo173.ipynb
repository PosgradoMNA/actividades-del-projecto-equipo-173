```python
cols = ["X10", "X13", "X17"]
summary={
    "Varianzas": df[cols].var(),
    "Valor Min": df[cols].min(),
    "Valor Max": df[cols].max()
}
pd.DataFrame(summary, index = cols).transpose()
```

```python
var_tot=df.var().sum()

var_X10=df.X10.var()
var_X13=df.X13.var()
var_X17=df.X17.var()

print('Varianza Total \t:',var_tot.round(2) )
print('Varianza X10 \t:', ((var_X10/var_tot)*100).round(2),'%' )
print('Varianza X13 \t:', ((var_X13/var_tot)*100).round(2),'%' )
print('Varianza X17 \t:', ((var_X17/var_tot)*100).round(2),'%' )
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
scaled=scaler.fit_transform(df)

scaled[:5]
```

```python
df_scaled = pd.DataFrame(scaled, columns=df.columns)
df_scaled.head()
```

```python
df_scaled.shape
```

```python
summary={
    "Varianzas": df_scaled[cols].var(),
    "Valor Min": df_scaled[cols].min(),
    "Valor Max": df_scaled[cols].max()
}
pd.DataFrame(summary, index = cols).transpose()
```

```python
from sklearn.decomposition import PCA

pcs=PCA()
pcs_t=pcs.fit_transform(df_scaled)
pcs_t[:5]
```

```python


df_pcs = pd.DataFrame({'% varianza explicada': np.round(pcs.explained_variance_ratio_,4) * 100,
                        '% varianza acumulada': np.cumsum(pcs.explained_variance_ratio_) * 100
})
df_pcs
```

```python
pcs_labels = [f'PC{i + 1}' for i in range(len(df_scaled.columns))]
df_pcs.index = pcs_labels
df_pcs
```

```python
df_pcs_tot = pd.DataFrame(pcs_t, columns =pcs_labels)
print("Varianza total variables originales: ", df_scaled.var().sum())
print("Varianza total de los componentes: ", df_pcs_tot.var().sum())
```
